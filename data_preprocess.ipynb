```python
sutherland_df = sutherland_df.set_index('Start Time')

hourly_df = sutherland_df.asfreq(freq = 'H')
hourly_df.head()
```

```python
sutherland_df = sutherland_df.reset_index()
hourly_df = hourly_df.reset_index()
hourly_df.head()
```

```python
print(len(hourly_df), '\n', '-------------')
print(hourly_df.dtypes)
```

```python
#append needed rows(if the last row has duration more than 60 min.) below the dataframe
last_dur = hourly_df.iloc[-1, 3]
while last_dur > 60.0:
    hourly_df = hourly_df.append(pd.Series(), ignore_index = True)
    last_dur = last_dur - 60.0
    
hourly_df.tail()
```

```python
hourly_df = hourly_df.assign(LGA = 'SUTHERLAND')
hourly_df = hourly_df.fillna(0)
hourly_df.head()
```

```python
#Check duration is more than 60 min. and fill to next row
for i in range(len(hourly_df)):
    if hourly_df.iloc[i, 3] > 60.0:
        print(i) #Check index
        duration_tmp = hourly_df.iloc[i, 3] #Duration  161
        
        #Copy the values of the row except 'Duration' when index is not the last row of the dataframe
        hourly_df.iloc[i+1, 2] = hourly_df.iloc[i, 2] #Customers Interrupted
        hourly_df.iloc[i+1, 4] = hourly_df.iloc[i, 4] #Reason

        hourly_df.iloc[i, 3] = 60.0
        hourly_df.iloc[i+1, 3] = duration_tmp - 60.0

        
hourly_df.tail()
```

```python
print('Total length is :', len(hourly_df))
print('\n')
print(hourly_df.dtypes)
```

```python
#Use pd.get_dummies to One-hot encoding 'Reason'

hourly_df_reason = pd.get_dummies(hourly_df.iloc[:, 4], prefix='Reason')
hourly_df = pd.concat([hourly_df, hourly_df_reason], axis=1)

#Rename no outage columns
hourly_df = hourly_df.rename(columns={'Reason_0': 'Reason_No outage'})

#Keep the One-Hot encoding 'Reason' and remove old 'Reason' column and remove 'LGA' column since the dataframe has only one location
hourly_df = hourly_df.drop(columns=['Reason'])
hourly_df = hourly_df.drop(columns=['LGA'])
```

```python
hourly_df.to_csv('sutherland-past-outages.csv', index=False)
hourly_df.head(10)
```

```python
outage_env = 0
outage_equip = 0
outage_3rd = 0
empty_row = 0

for i in range(len(hourly_df)):
    if hourly_df.iloc[i]['Reason_Environmental'] != 0:
        outage_env = outage_env + 1
        continue
    elif hourly_df.iloc[i]['Reason_Equipment fault'] != 0:
        outage_equip = outage_equip + 1
        continue
    elif hourly_df.iloc[i]['Reason_Third party'] != 0:
        outage_3rd = outage_3rd + 1
        continue
    else:
        empty_row = empty_row + 1
        
print('Number of total row is  : ', len(hourly_df), '\n')

print('Number of Outage row is : ', outage_env + outage_equip + outage_3rd)
print('Number of Normal row is : ', empty_row, '\n')

print('Propotion of Outage is  : ', (outage_env + outage_equip + outage_3rd) / len(hourly_df))
print('Propotion of Normal is  : ', empty_row / len(hourly_df), '\n')

print('Outage cause by is environmental   : ', outage_env)
print('Outage cause by is equipment fault : ', outage_equip)
print('Outage cause by is third party     : ', outage_3rd)
```
